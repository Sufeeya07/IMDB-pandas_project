# 20. Classify Movies Based on Ratings [Excellent, Good and Average]

```python
df.head(2)
```

```python
def rating(rate):
    if rate>7.0:
        return 'Excellent'
    elif rate>=6.0:
        return 'Good'
    else:
        return "Average"

r=df['Rating'].apply(rating)

# create new col rating_cat
df['Rating_cat']=r
```

# 22. Find Unique Values From Genre

```python
df['Genre']
```

```python
list=[]
for value in df['Genre']:
    list.append(value.split(','))
list
```

```python
one_d=[]
for item in list:
    for iteml in item:
        one_d.append(iteml)
one_d
```

```python
uni_list=[]
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
uni_list
```
